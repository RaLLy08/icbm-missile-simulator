```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
#### ROCKET THRUST EQUATION

rocket_payload_mass = 1000  # kg
rocket_fuel_mass = 28400
# Parameters
m0 = rocket_fuel_mass       # Initial mass (kg)
mf = rocket_payload_mass       # (kg) after fuel burn
ve = 3000.0       # Exhaust velocity (m/s)
dmdt = 50.0        # (kg/s)
F = dmdt * ve     # (N)

# Time setup
burn_time = (m0 - mf) / dmdt  # Total burn time (s)
t = np.linspace(0, burn_time, 1000)  

# Mass as a function of time
m_t = m0 - dmdt * t 

# Acceleration (a = F / m(t))
a_t = F / m_t

# Velocity (integrate acceleration: v = a(t) * dt)
v_t = np.cumsum(a_t) * (t[1] - t[0])  # Approximate integral

# Plot
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 8))

ax1.plot(t, m_t, 'b-')
ax1.set_ylabel('Mass (kg)')
ax1.grid()
ax2.plot(t, a_t, 'r-')
ax2.set_ylabel('Acceleration (m/s)')
ax2.grid()
ax3.plot(t, v_t, 'g-')
ax3.set_xlabel('Time (s)')
ax3.set_ylabel('Velocity (m/s)')
ax3.grid()
plt.suptitle('Rocket Simulation: Mass, Acceleration, and Velocity vs. Time')
plt.tight_layout()
plt.show()
```

```python
delta_v_theory = ve * np.log(m0 / mf)
print(f"Tsiolkovsky v: {delta_v_theory:.2f} m/s")
print(f"Simulated v: {v_t[-1]:.2f} m/s")
```
